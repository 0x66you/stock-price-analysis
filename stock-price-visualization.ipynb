```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta
today = date.today()

end_date = today.strftime("%Y-%m-%d")
# print(end_date) 
start_date = date.today() - timedelta(days=360)
start_date = start_date.strftime("%Y-%m-%d")
# print(start_date)
data = yf.download('2303.TW',
                    start=start_date,
                    end=end_date,
                    progress=False)
print(data.tail())
```

```python
import plotly.express as px
figure = px.line(data, x = data.index,
                y = "Close",
                title = "收盤價格折線圖(2303聯電)")
figure.show()
```

```python
import plotly.graph_objects as go
figure = go.Figure(data=[go.Candlestick(
    x = data.index,
    open = data["Open"],
    high = data["High"],
    low = data["Low"],
    close = data["Close"]
)])
figure.update_layout(title="K線(2303聯電)")
figure.show()
```

```python
figure = px.bar(data,x=data.index,
                y = "Close",
                title = "長條圖(2303聯電)")
figure.show()
```

```python
figure = px.line(data,x = data.index,
                y='Close',
                range_x=['2021-07-01','2021-12-31'],
                title="指定區間折線圖(聯電)")
figure.show()
```

```python
figure = go.Figure(data = [go.Candlestick(x = data.index,
                                        open = data["Open"], 
                                        high = data["High"],
                                        low = data["Low"], 
                                        close = data["Close"])])
figure.update_layout(title = "k線+區間軸+按鈕(2303聯電)")

figure.update_xaxes(
    rangeslider_visible = True,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = "1m", step = "month", stepmode = "backward"),
            dict(count = 6, label = "6m", step = "month", stepmode = "backward"),
            dict(count = 1, label = "YTD", step = "year", stepmode = "todate"),
            dict(count = 1, label = "1y", step = "year", stepmode = "backward"),
            dict(step = "all")
        ])
    )
)
figure.show()
```
